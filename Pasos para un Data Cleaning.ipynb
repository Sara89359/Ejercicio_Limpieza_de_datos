```python
# Imagina que trabajas con un gimnasio que te envía un archivo CSV llamado . El archivo contiene registros de entrenamientos, pero está lleno de problemas típicos: duplicados, fechas mal formateadas, valores nulos y algunos errores de entrada.
# Tu misión es limpiar el dataset y obtener un pequeño análisis final.
```

```python
# PASO 1: Hay que importar la libreria PANDAS para poder usarla durante el ejercicio y ver el archivo en este ejercicio.

import pandas as pd

print('Pandas listo, versión:', pd.__version__)
```

```python
# PASO 2: Ahora tenemos que ver la tabla en pantalla. Para ello, usamos el metodo "pd.read_csv": 

tabla = "gimnasio_clientes.csv"
datos = pd.read_csv(tabla)
datos
```

```python
# PASO 3: ¿Y que datos nos han dado? ¿Que tenemos que analizar?

datos.dtypes
```

```python
# PASO 4: Sabiendo que tipo de datos tenemos, vamos a mirar la tabla para ver si hay algún dato que nos llame la atención:

    # En este caso, llama la atencion las ' despues de las fechas, datos anomalos en la duracion y en las calorias. 
```

```python
# PASO 5: ¿Cuantos entrenamientos tenemos en total? --> Tenemos que saber de donde partimos.

print(f"En total, con datos invalidos, tenemos una grafica del tamaño de: {datos.shape} (filas,columnas)")
```

```python
# PASO 6: Ahora vamos a sacar estadísticas para comprobar si algo más nos llama la atención:

datos.describe()
```

```python
# Lo que nos indica que:
    # Hay 8 datos validos de la columna de duración y 7 en la columna de calorias --> Sobra un dato de la primera columna.
    # Tenemos un minimo de -20 en la duración.
    # Tenemos un dato muy extremo de 4 calorias y otro de 7 calorias. Eso se hace pestañeando, asi que lo voy a considerar error. 
    # Tenemos un dato de una duración de 282 minutos entrenando (son unas 5h entrenando). También es un valor muy extremo. 
```

```python
# PASO 7: Vamos a empezar a analizar la tabla, ya en serio. Lo primero, le vamos a decir que busque datos nulos.

datos.isnull().sum()
```

```python
# PASO 8: Tenemos un dato nulo en la columna de calorias. Como no nos interesa trabajar con él, vamos a eliminarlo de la tabla:

    # Primero, tenemos que localizar en que fila se encuentra ese dato nulo:

datos[datos.isnull().any(axis=1)]

# Aclaración: 
    #  Con el comando -datos.isnull()- estás diciendo: "Crea un DataFrame de True/False indicando dónde hay nulos"
    #  Con el comando -any(axis=1)- estás diciendo: "Para cada fila, dime si alguna de sus columnas tiene un valor nulo"
    #  Con los dos juntos, estas diciendo: "Revisa cada fila (por eso axis=1) y devuelve True si esa fila tiene al menos un nulo"

```

```python
    # Segundo, ya sabemos que está en la fila 4. Ahora vamos a cargarnosla:

datos = datos.dropna(subset=['Calories'])
```

```python
    # Tercero y último, vamos a reiniciar el dataset sin ese valor nulo:

datos = datos.reset_index(drop=True)
```

```python
    # Y claro, darle a imprimir para ver que tenemos:

print('Filas restantes:', len(datos))
print('Nulos totales:', datos.isnull().sum().sum())
datos
```

```python
# PASO 9: Ahora que tenemos la tabla sin valores nulos, vamos a corregir las fechas, poniendo todas en el mismo formato:

    # En primer lugar, vamos a sacar toda la columna de "Fechas":

datos['Date'].tolist()
```

```python
    # En segundo lugar, vamos a eliminar las comillas que nos sobren de las fechas:

datos['Date'] = (
    datos['Date']
    .str.replace("'", '', regex=False)
)
print(datos['Date'].tolist())
```

```python
# Para asegurarnos, volvemos a llamar a la tabla a ver como queda:

datos
```

```python
# PASO 10: Ahora que tenemos la tabla limpia, vamos a buscar duplicados:

datos[datos.duplicated(keep=False)]
```

```python
# Tenemos 2 duplicados! Vamos a eliminar uno de ellos de la lista:

datos = datos.drop_duplicates()
```

```python
# ¿Como ha quedado la tabla ahora que hemos borrado otro dato?

datos
```

```python
# PASO 11: Vamos a poner todas las fechas en el mismo formato para manejar mejor los datos:

datos['Date'] = pd.to_datetime(datos['Date'], dayfirst=True, format='mixed')

print(datos['Date'])
```

```python
# PASO 12: Seguimos teniendo datos anómalos en la columna de "duration", ¿Que hacemos con ellos?

    # Lo primero es localizar los valores extremos que pasen de 2h de entrenamiento:

datos[datos['Duration'] > 120]
```

```python
    # Este dato, posiblemente sea un error, ya que no pueden ser 4 calorias en 282 minutos. Como ambos valores están mal, los voy a eliminar también.

datos = datos[datos['Duration'] <= 120]

    # Con esta formula, mantiene en la tabla los resultados que entren dentro de la condición
```

```python
    # Comprobamos como queda la tabla: 

datos
```

```python
    # Vamos a buscar los resultados que estén por debajo de 10 minutos (un entrenamiento de menos de 10 minutos NO es un entrenamiento):

datos[datos['Duration'] < 10]
```

```python
    # En este caso, vamos a cambiar la duración de -20 a 20. 

datos['Duration'] = datos['Duration'].replace(-20, 20)
```

```python
    # Volvermos a sacar la tabla para comprobar que está corregido:

datos
```
