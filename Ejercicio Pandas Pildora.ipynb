## 🎯 Paso 3 — Eliminar duplicados

Parece que algún entrenamiento se registró dos veces.

### Ejercicio 3.1 — Encontrar los duplicados
Muestra las filas que están repetidas.

> 💡 **Pista:** `df[df.duplicated(keep=False)]`  
> `keep=False` marca **todas** las copias como duplicadas

```python
# ✏️ Muestra las filas duplicadas
df[df.duplicated(keep=False)]
```

### Ejercicio 3.2 — Eliminar los duplicados
Quédate solo con la primera aparición y reinicia el índice.

> 💡 **Pistas:**
> - `.drop_duplicates()` elimina duplicados
> - `.reset_index(drop=True)` reinicia el índice
> - Recuerda guardar: `df = ...`

```python
# ✏️ Elimina duplicados y reinicia el índice
df_nueva = df.drop_duplicates()
df_nueva = df_nueva.reset_index(drop=True)

print('Filas después de eliminar duplicados:', len(df_nueva))
```

---
## 🕳️ Paso 4 — Manejar valores nulos

Tenemos valores nulos en `Date` y `Calories`. Los trataremos de forma diferente.

### Ejercicio 4.1 — Ver las filas con nulos
Muestra solo las filas que tienen algún valor nulo.

> 💡 **Pista:** `df[df.isnull().any(axis=1)]`  
> `axis=1` significa "busca en cada fila"

```python
# ✏️ Muestra las filas con nulos
df[df.isnull().any(axis=1)]
```

### Ejercicio 4.2 — Rellenar Calories con la media
Rellena los huecos de `Calories` con la media de esa columna, redondeada a 1 decimal.

> 💡 **Pistas:**
> - Calcula primero: `df['Calories'].mean()`
> - Redondea: `round(media, 1)`
> - Rellena: `.fillna(valor)`

```python
# ✏️ Calcula la media e imprímela
media = df_nueva['Calories'].mean()
media = round(media, 1)
# ✏️ Rellena los huecos de Calories

df_nueva['Calories'] = df_nueva['Calories'].fillna(media)
df_nueva['Calories'] = df_nueva['Calories'].round(1)

print('Nulos en Calories:', df_nueva['Calories'].isnull().sum())
```

### Ejercicio 4.3 — Eliminar la fila con Date nula
La fila que no tiene fecha no nos sirve. Elimínala y reinicia el índice.

> 💡 **Pistas:**
> - `.dropna(subset=['Date'])` elimina filas con nulo en Date
> - `.reset_index(drop=True)`

```python
# ✏️ Elimina la fila sin fecha
df_nueva = df_nueva.dropna(subset=['Date'])
df_nueva = df_nueva.reset_index(drop=True)
print('Filas restantes:', len(df_nueva))
print('Nulos totales:', df_nueva.isnull().sum().sum())
```

---
## 📅 Paso 5 — Limpiar la columna de fechas

La columna `Date` tiene un problema: la mayoría de fechas tienen comillas (`'2020/12/01'`) y **una** tiene un formato completamente diferente (`20201226`).

### Ejercicio 5.1 — Ver los valores únicos de Date
Imprime todos los valores de la columna `Date` para ver el problema.

> 💡 **Pista:** `df['Date'].tolist()`

```python
# ✏️ Muestra todos los valores de Date
df['Date'].tolist()
```

### Ejercicio 5.2 — Eliminar las comillas
Elimina las comillas simples (`'`) que rodean las fechas.

> 💡 **Pista:** `.str.replace("'", '', regex=False)`

```python
# ✏️ Elimina las comillas de la columna Date
df_nueva['Date'] = (
    df_nueva['Date']
    .astype(str)
    .str.strip()
    .str.replace('"', '', regex=False)
    .str.replace("'", '', regex=False)
)
print(df_nueva['Date'].tolist())
```

### Ejercicio 5.3 — Convertir al tipo fecha
Convierte `Date` al tipo de dato `datetime` de pandas.

> 💡 **Pista:** `pd.to_datetime(df['Date'], infer_datetime_format=True)`  
> Con `infer_datetime_format=True` pandas intenta detectar el formato solo

```python
# ✏️ Convierte Date a datetime
df_nueva['Date'] = pd.to_datetime(df_nueva['Date'], dayfirst=True, format='mixed')

print(df_nueva['Date'])
```

### Ejercicio 6.4 — Eliminar la fila incorrecta
Esa fila tiene datos imposibles. Elimínala y reinicia el índice.

> 💡 **Pista:** quédate con las filas donde `Maxpulse >= Pulse`  
> `df = df[df['Maxpulse'] >= df['Pulse']].reset_index(drop=True)`
